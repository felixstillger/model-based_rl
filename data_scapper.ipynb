```python
from bs4 import BeautifulSoup
import requests


# http://jobshop.jjvh.nl/instance.php?instance_id=2

#print(soup.prettify())
```

```python
import pandas as pd
contents=pd.DataFrame
dic={}
for _ in range(1,243):
    r = requests.get('http://jobshop.jjvh.nl/instance.php?instance_id='+str(_))
    soup = BeautifulSoup(r.text, 'html.parser')
    mydivs = soup.find_all("td", {"class": "basicinfo_title"})
    mytds=soup.find_all("td",{ "class":'basicinfo_desc'})
    name=""
    for div,div2 in zip(mydivs,mytds):
        ID=div.get_text(strip=True)
        VALUE=div2.get_text(strip=True)
        if ID=="Name":
            name=VALUE
            dic[name,ID]=name

        else:
            dic[name,ID]=VALUE 


    

```

```python

import pandas as pd
df = pd.DataFrame(dic.values(), index=pd.MultiIndex.from_tuples(dic.keys())).unstack().T.droplevel(level=0).T
display(df)
entry=df['Number of jobs']['abz5']

```
